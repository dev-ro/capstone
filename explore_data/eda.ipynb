```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
filepath = '../data/processed/'
filename = 'movies.tsv'
movies = pd.read_csv(filepath + filename, sep='\t')
```

```python
movies.columns
```

```python
# in order to better define a 'good' movie, we need to look at the distribution of the ratings
# and see where the majority of the ratings lie

plt.hist(movies['averageRating'], bins=50)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.xlim(0, 10)
plt.show()
```

```python
# From the top 20000 most rated movies on IMDB
# The distribution of ratings is right-skewed, with the majority of the ratings between 5 and 8.
# For now, wee can define a 'good' movie as one with a rating greater than 7.

good_movies = movies[movies['averageRating'] > 7]
good_movies

# export to tsv
good_movies.to_csv(filepath + 'good_movies.tsv', sep='\t', index=False)
```
